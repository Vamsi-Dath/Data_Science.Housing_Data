```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("Zip(zhvf_and_zhvi)_combined/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")
df1.head(10)       
```

```python
df_filtered = df1.loc[:, ['StateName', 'City']].join(df1.iloc[:, 9:])
df_filtered
```

```python
# Apply row-wise median filling starting from the 2nd column onwards
df_filtered.iloc[:, 2:] = df_filtered.iloc[:, 2:].apply(lambda row: row.fillna(row.median()), axis=1)
```

```python
df_filtered
```

```python
df_filtered.isnull().sum()
```

```python
df_long = df_filtered.melt(id_vars=['StateName', 'City'], var_name='Date', value_name='Price')

df_long['Date'] = pd.to_datetime(df_long['Date'], errors='coerce')

df_long
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'

# Apply the function to create a new 'Season' column
df_long['Season'] = df_long['Date'].dt.month.apply(get_season)

# Display the DataFrame to verify the new column
df_long.tail()
```

```python
import matplotlib.pyplot as plt

# Ensure 'Date' is in datetime format and extract the year for grouping
df_long['Year'] = df_long['Date'].dt.year

# Group by 'Season' and 'Year' and calculate the average price for each season per year
seasonal_trends = df_long.groupby(['Season', 'Year'])['Price'].mean().reset_index()

# Pivot the data to have years on the x-axis and seasons as separate lines
seasonal_trends_pivot = seasonal_trends.pivot(index='Year', columns='Season', values='Price')

# Plotting the seasonal trends
plt.figure(figsize=(12, 6))
for season in seasonal_trends_pivot.columns:
    plt.plot(seasonal_trends_pivot.index, seasonal_trends_pivot[season], marker='o', label=season)

plt.title('Average House Prices by Season Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.legend(title="Season")
plt.grid(True)
plt.show()
```
